The data is the fuel quality data from the Federal Energy Regulatory Commission which is provided by the United States Energy Information Administration. The data consists of the following columns:

'Record_id' : record id

'Utility_id_ferc1': Utility id assigned by the FERC

'Report_year': year of report

'Plant_name_ferc1': the name of the plant

'Fuel_type_code_pudl': the type of fuel

'Fuel_unit': the unit of fuel

'Fuel_qty_burned': the quantity of fuel burned

'Fuel_mmbtu_per_unit': the measure of energy per unit

'fuel_cost_per_unit_burned': the fuel cost per unit burned

'Fuel_cost_per_unit_delivered': the cost of fuel delivered per unit

'fuel_cost_per_mmbtu': the cost of fuel per mmbtu

```python
#Import pandas library
import pandas as pd
```

```python
#Load Dataset
url='https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true'
fuel_data = pd.read_csv(url, error_bad_lines=False)
```

```python
#Summary Statistics
fuel_data.describe(include='all')
```

```python
#check for missing values
fuel_data.isnull().sum()
```

```python
#Percentage of missing value
percent=fuel_data.isnull().sum()/len(fuel_data)
percent*100
```

```python
#use groupby to count the sum of each unique value in the fuel unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
fuel_data[['fuel_unit']] = fuel_data[['fuel_unit']].fillna(value='mcf')
```

```python
#check if missing values have been filled
fuel_data.isnull().sum()
```

```python
fuel_data.groupby('fuel_type_code_pudl').mean()
```

```python
#Skewness and Kurtosis of fuel quantity burned
from scipy.stats import skew
fuel_data['fuel_qty_burned'].skew()
```

```python
fuel_data['fuel_qty_burned'].kurtosis()
```

```python
fuel_data.corr()
```

```python
#Which year has highest average fuel cost per unit delivered
fuel_data.groupby('report_year')['fuel_cost_per_unit_delivered'].mean()
```
